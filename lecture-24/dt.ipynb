```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../data/titanic.csv")
```

```python
data
```

```python
data.columns
```

```python
selected = data[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
```

```python
selected
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
selected["Sex"] = LabelEncoder().fit_transform(selected["Sex"])
```

```python
selected
```

```python
selected.describe()
```

```python
clean = selected.dropna()
```

```python
dt = clean.groupby(['Pclass', 'Sex']).mean().reset_index()
```

```python
dt
```

```python
def fill_age(row):
    if pd.isna(row["Age"]):
        row["Age"] = dt.loc[(dt["Pclass"] == row["Pclass"]) & (dt["Sex"] == row["Sex"])].iloc[0, 3]
    return row 

filled = selected.apply(fill_age, axis=1)
```

```python
filled.describe()
```

```python
filled.columns
```

```python
X = filled[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
```

```python
y = filled['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
