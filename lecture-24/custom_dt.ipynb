```python
clean = selected.dropna()
```

```python
dt = clean.groupby(['Pclass', 'Sex']).mean().reset_index()
```

```python
def fill_age(row):
    if pd.isna(row["Age"]):
        row["Age"] = dt.loc[(dt["Pclass"] == row["Pclass"]) & (dt["Sex"] == row["Sex"])].iloc[0, 3]
    return row 

filled = selected.apply(fill_age, axis=1)
```

```python
X = filled[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
```

```python
y = filled['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
class Node:
    
    def __init__(self, label=None, thresh=None, result=None):
        self.label = label
        self.thresh = thresh
        self.result = result


class CustomDT:
    
    def __init__(self, max_depth=5):
        self.max_depth = max_depth
        
    @classmethod
    def entropy(self, data):
        uniques, counts = np.unique(data, return_counts=True)
        probs = counts / len(data)
        return -np.sum(probs * np.log(probs))
    
    @classmethod
    def infogain(self, X, y, label):
        
        thresh = X[label].mean()
        
        left = y[X[label] < thresh]
        right = y[X[label] >= thresh]
        
        ent = CustomDT.entropy(y)
        
        lefte = CustomDT.entropy(left)
        righte = CustomDT.entropy(right)
        
        pl = len(left) / len(y)
        pr = len(right) / len(y)
        
        return ent - (pl*lefte + pr * righte)
        
    def fit(self, X, y):
        self.__root = self.__rec_fit(X, y, self.max_depth)
        
    def __rec_fit(self, X, y, depth):
        
        if depth == 0:
            return Node(result=y.mean())
        
        gains = []
        
        for col in X.columns:
            gain = CustomDT.infogain(X, y, col)
            gains.append([gain, col])
        
        gains = sorted(gains)
        best_gain, best_label = gains[-1]
        
        if best_gain <= 0:
            return Node(result=y.mean())
        else:
            thresh = X[best_label].mean()
            select_left = X[best_label] < thresh
            select_right = X[best_label] >= thresh
            
            X_left = X.loc[select_left]
            X_right = X.loc[select_right]
            
            y_left = y[select_left]
            y_right = y[select_right]
            
            node = Node(label=best_label, thresh=thresh)
            
            node.left = self.__rec_fit(X_left, y_left, depth-1)
            node.right = self.__rec_fit(X_right, y_right, depth-1)
            
            return node
    
    def predict_point(self, row, node):
        
        if node.label == None:
            return node.result
        
        if row[node.label] < node.thresh:
            return self.predict_point(row, node.left)
        else:
            return self.predict_point(row, node.right)
        
    def predict(self, X):
        
        results = []
        
        for idx, row in X.iterrows():
            results.append(self.predict_point(row, self.__root))
        
        return (np.array(results) > .5).astype(int)
            
    def score(self, X, y):
        
        yp = self.predict(X)
        
        return (np.array(y) == yp).mean()
            
```

```python
model = CustomDT()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
