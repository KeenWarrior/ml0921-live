```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = Image.open("./strange.jpeg")
```

```python
npimg = np.array(img)
```

```python
npimg.shape
```

```python
X = npimg.reshape([-1, 3])
```

```python
X.shape
```

```python
model = KMeans(4)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
pallets
```

```python
plt.imshow(pallets.reshape([1, -1, 3]))
```

```python
img
```

```python
X_mod = pallets[model.labels_]
```

```python
X_mod.shape
```

```python
mod_img = X_mod.reshape(npimg.shape)
```

```python
plt.imshow(mod_img)
```
