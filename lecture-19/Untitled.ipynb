```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
img = Image.open("../data/monastery.jpeg")
```

```python
im = np.array(img)
```

```python
im = cv2.resize(im, (400, 400))
```

```python
def applyFilter(im, fil):
    k = len(fil)
    fil = fil.reshape([fil.shape[0], fil.shape[1], 1])
    target = np.empty([im.shape[0]-k+1, im.shape[1]-k+1, 3])

    for r in range(target.shape[0]):
        for c in range(target.shape[1]):
            cut = im[r:r+k, c:c+k]
            target[r, c] = (cut*fil).mean(axis=0).mean(axis=0)

    return np.abs(target.astype(int))
```

```python
plt.imshow(applyFilter(im, np.array([list(range(-2, 3))]*5).T))
```

```python
plt.imshow(applyFilter(im, np.array([list(range(-2, 3))]*5)))
```

```python
plt.imshow(applyFilter(im, np.ones([7, 7])))
```
